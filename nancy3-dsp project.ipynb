```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
```

```python
df["Sex"]=encoder.fit_transform(df["Sex"])
df["Sex"]
```

```python
df["BP"]=encoder.fit_transform(df["BP"])
df["BP"]
```

```python
df["Cholesterol"]=encoder.fit_transform(df["Cholesterol"])
df["Cholesterol"]
```

```python
df["Drug"]=encoder.fit_transform(df["Drug"])
df["Drug"].tail(10)
```

```python
features=["Age","Sex","BP","Cholesterol","Na_to_K"]
x=df[features]
x
```

```python
y=df["Drug"]
y
```

```python
from sklearn.model_selection import train_test_split 
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=401)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg=LogisticRegression()
logreg.fit(x_train,y_train)
y_pred=logreg.predict(x_test)
y_pred
```

```python
print(logreg.score(x_train,y_train))
print(logreg.score(x_test,y_test))
```
